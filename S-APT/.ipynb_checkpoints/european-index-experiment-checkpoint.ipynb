## Credit (CDS spread)

```python
# placeholder for sovereign CDS, using CDS spread as of 02/16/2021
tmp=read_data(os.path.join(DATA_DIR, 'msci_px_to_book_ratio.csv'), rename_mapping)
tmp['Austria']=11.38
tmp['Belgium']=14.46
tmp['Finland']=12.52
tmp['France']=17.05
tmp['Germany']=11.65
tmp['Greece']=78.60
tmp['Ireland']=16.15
tmp['Italy']=71.42
tmp['Netherlands']=11.50
tmp['Portugal']=33.53
tmp['Spain']=36.84
credit=Sort_Portfolio(tmp)
credit.plot_char()
```

```python
credit.form_portfolio(px,num_extreme=2, lookback=dt.timedelta(days=1), lookforth_period_in_price=150)
credit_long_short=credit.get_portfolio()
print(credit_long_short.mean()/credit_long_short.std())
credit_long_short.describe()
```

So basically the strategy is to long the 2 most credit-worthy countries: Austria, Netherland and Germany, and short the two least credit-worthy: Greece and Italy. Hold the long-short for 150 days.



```python
credit.plot_portfolio()
```

- Probably with dynamic CDS data, it can still improve on Sharpe ratio
- The longest drawdown is in 2014 (less than a year), which is a result of the Greek outperformance, which in turn is due to Greece accepting the bailout after referendum and back on the bond market.



## Size

```python
size_file=os.path.join(DATA_DIR, 'msci_indx_market_cap.csv')
size=Sort_Portfolio(read_data(size_file, rename_mapping))
size.plot_char(logy=True)
```

```python
size.form_portfolio(px, num_extreme=1, lookback=dt.timedelta(days=10), lookforth_period_in_price=150)
size_long_short=size.get_portfolio()
print(size_long_short.mean()/size_long_short.std())
size_long_short.describe()
```

The strategy is longing the country with smallest stock market, which could be either Spain, Portugal or Greece, and shorting the one with biggest stock market, which is quite stably France. Hold the long-short for 150 days.



```python
size.plot_portfolio()
```

- Consistent with the literature, size factor is not very robust.
- In particular, its performance may suffer precisely due to susception of 'junks', as described in [Size matters, if you control your junk](https://www.evernote.com/shard/s191/nl/21353936/a1cdd686-8ab5-3903-de9a-f3649248cc89?title=Size%20matters,%20if%20you%20control%20your%20junk).

## Momentum

```python
simple_return=px.pct_change()
negative_log_return=(-1)*np.log(simple_return.fillna(value=0)+1) ## we need the negative sign since in the long-short it is bottom-top
momentum=Sort_Portfolio(negative_log_return)
momentum.plot_char()
```

```python
momentum.form_portfolio(px, num_extreme=1, lookback=dt.timedelta(days=180), lookforth_period_in_price=220)
momentum_long_short=momentum.get_portfolio()
print(momentum_long_short.mean()/momentum_long_short.std())
momentum_long_short.describe()
```

Long the 1 winner and short 1 loser


```python
momentum.plot_portfolio()
```

- The momentum strategy seems to do more decent than other factors, but suffer from crashes from time to time, probably because regime change is common during this period. As such, the vol of this strategy is also high.
- Interestingly, as mentioned in the literature, there is no need to devise a one-month or so 'cooling period' for momentum.



## Equal-Weight

```python
simple_return=px.pct_change(periods=220).dropna()*100/11
equal_weight=simple_return.sum(axis=1)
equal_weight['zero']=0
equal_weight.plot(figsize=(2*10,10))
plt.grid()
plt.show()
```

- Equal-weight can be viewed as a variant of the size factor, but there is no long-short, so it loads on the overall stock market and thus crashed in the aftermaths of the financial crisis. This is particularly dangerous considering the co-integration of stock returns.
- Despite its drawbacks above, it still has a positive and higher Sharpe than the size factor.



```python
equal_weight.describe()
```

## Combined

```python
combined=(1.5*credit_long_short+size_long_short+0.3*momentum_long_short).dropna()
combined['zero']=0
combined.plot(figsize=(2*10,10))
print(combined.mean()/combined.std())
combined.describe()
plt.grid()
plt.show()
```

It should look like this. 

- The combined portfolio is an improvement to both the credit and momentum portfolio: smaller and shorter drawdowns and higher Sharpe. But it seems since all strategies do badly in 2014, the combined one is no exception.
- The payoff of the combined one is smoother than the momentum one in that the latter comes in and out of profit/loss, so a size strategy using lagged payoff can further smooth PnL. The involvement of size also shorten the duration and magnetude of the drawdown in 2014.
