## Momentum

```python
simple_return=px.pct_change()
negative_log_return=(-1)*np.log(simple_return.fillna(value=0)+1) ## we need the negative sign since in the long-short it is bottom-top
momentum=Sort_Portfolio(negative_log_return)
momentum.plot_char()
```

(It should look like this)



```python
momentum.form_portfolio(px, num_extreme=1, lookback=dt.timedelta(days=180), lookforth_period_in_price=220)
momentum_long_short=momentum.get_portfolio()
print(momentum_long_short.mean()/momentum_long_short.std())
momentum_long_short.describe()
```

## Combined

```python
combined=(1.5*credit_long_short+size_long_short+0.3*momentum_long_short).dropna()
combined['zero']=0
combined.plot(figsize=(2*10,10))
print(combined.mean()/combined.std())
combined.describe()
plt.grid()
plt.show()
```

It should look like this. 

- The combined portfolio is an improvement to both the credit and momentum portfolio: smaller and shorter drawdowns and higher Sharpe. But it seems since all strategies do badly in 2014, the combined one is no exception.
- The payoff of the combined one is smoother than the momentum one in that the latter comes in and out of profit/loss, so a sizing strategy using lagged payoff can further smooth PnL.
